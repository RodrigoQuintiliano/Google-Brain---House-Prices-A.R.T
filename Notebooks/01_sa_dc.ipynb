```python
from pathlib import Path
import pandas as pd
```

```python
REPO_ROOT = Path.cwd().parent
TRAIN_IN = REPO_ROOT / "Data" / "Processed" / "train_ohe_knn.csv"
TEST_IN  = REPO_ROOT / "Data" / "Processed" / "test_ohe_knn.csv"
```

```python
train = pd.read_csv(TRAIN_IN)
test  = pd.read_csv(TEST_IN)
```

```python
assert "SalePrice" in train.columns
assert "SalePrice" not in test.columns
```

```python
X = train.drop(columns=["SalePrice"])
```

```python
X_cols = set(X.columns)
T_cols = set(test.columns)
print("Mismatch columns (should be 0):", len(X_cols ^ T_cols))
```

```python
X = X.rename(columns=lambda c: c.replace(" ", "_"))
test = test.rename(columns=lambda c: c.replace(" ", "_"))
```

```python
train_fixed = X.copy()
train_fixed["SalePrice"] = train["SalePrice"]
```

```python
TRAIN_OUT = REPO_ROOT / "Data" / "Processed" / "train_ohe_knn_fixed.csv"
TEST_OUT  = REPO_ROOT / "Data" / "Processed" / "test_ohe_knn_fixed.csv"
```

```python
train_fixed.to_csv(TRAIN_OUT, index=False)
test.to_csv(TEST_OUT, index=False)
```

```python
print("Saved:", TRAIN_OUT)
print("Saved:", TEST_OUT)
print("Final shapes:", train_fixed.shape, test.shape)
```
